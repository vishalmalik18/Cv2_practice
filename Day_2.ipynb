```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("dog.jpeg")
cv2.imshow("dog image",image)
cv2.destroyAllWindows()
cv2.destroyAllWindows()
```

```python
def viewimage(image,name_of_window):
    cv2.namedWindow(name_of_window,cv2.WINDOW_NORMAL)
    cv2.imshow(name_of_window,image)
    cv2.waitKey()
    cv2.destroyAllWindows()
```

```python
image_1 = cv2.imread("dog.jpeg")
viewimage(image_1,"dog image")
```

```python
scale_percent = 30
width = int(image_1.shape[1]*scale_percent/100)
height = int(image_1.shape[0]*scale_percent/100)
dim = (width,height)
resized_image = cv2.resize(image_1,dim,interpolation=cv2.INTER_AREA)
viewimage(resized_image,"dog image")
```

```python
(h,w,d) = image_1.shape
center = (w//2,h//2)
M = cv2.getRotationMatrix2D(center,180,1.0)
rotated = cv2.warpAffine(image_1,M,(w,h))
viewimage(rotated,'dog image')
```

```python
(h,w,d) = image_1.shape
center = (w//2,h//2)
M = cv2.getRotationMatrix2D(center,90,1.0)
rotated = cv2.warpAffine(image_1,M,(w,h))
viewimage(rotated,'dog image')
```

```python
face_identifier=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
image=cv2.imread("opencv.jpg")
faces=face_identifier.detectMultiScale(image)
if faces is ():
    print("NO human faces found")
else: 
    for (x,y,w,h) in faces:
        cv2.rectangle(image,(x,y),(x+w,y+h),(0,0,255),2)
        cv2.imshow("Face Detection",image)
        cv2.waitKey()
    cv2.destroyAllWindows()
```
